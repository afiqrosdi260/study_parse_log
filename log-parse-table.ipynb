```python
import parse as pr
import pandas as pd

simple_log_data = [
    "INFO [2023-05-17 12:34:56.789] This is a simple log message",
    "ERROR [2023-05-18 01:23:45.678] An error occurred",
    "WARNING [2023-05-19 10:20:30.123] This is a warning message"
]

formatt="{type} [{date}] {message}" 

collection=[]
for line in simple_log_data:
    parsed=pr.parse(formatt,line).named
    collection.append(parsed)

pd.DataFrame(collection) 
```

### Testing using re n pandas

```python
import pandas as pd
import parse
import re
```

```python
# Define the log format with a regular expression
log_regex = re.compile(
    r'^(?P<timestamp>\S+)\s+(?P<duration>\S+)\s+(?P<type>\S+)\s+(?P<details>.+)$'
)

# Initialize a list to store parsed log entries
log_collection = []
log_file_path = 'LCS0001.LOG'

# Read the log file
with open(log_file_path, 'r') as log_file:
    log_file_lines = log_file.readlines()

# Process each log line
for line in log_file_lines:
    match = log_regex.match(line)
    if match:
        # If the line matches the regex, extract data into a dictionary
        log_data = match.groupdict()
        log_collection.append(log_data)
    else:
        print(f"Line does not match format: {line}")

# Convert to Pandas DataFrame
df = pd.DataFrame(log_collection)

# Show the DataFrame
df
```

```python
# Filter rows where the 'type' column is 'REC', 'RES', or 'REM'
df_filtered = df[df['type'].isin(['REC', 'RES', 'REM', 'APR'])]

# Sort the DataFrame by 'timestamp' (or any other column you prefer)
df_filtered_sorted = df_filtered.sort_values(by='timestamp')

# Display the resulting filtered and sorted DataFrame
df_filtered_sorted
```
